```python
import re
from pathlib import Path

train_py_path = Path("train.py")
content = train_py_path.read_text()

# Target the exact block where load_dataset runs
old_block = """    from datasets     import load_dataset
    from transformers import GPT2TokenizerFast

    os.makedirs(cfg.data_dir, exist_ok=True)
    tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")
    dataset   = load_dataset("wikitext", "wikitext-2-raw-v1")

    def encode_split(split_name: str) -> int:
        hf_split = "validation" if split_name == "val" else split_name
        texts    = "\\n\\n".join(
            t for t in dataset[hf_split]["text"] if t.strip()
        )
        ids = tokenizer.encode(texts)"""

# Swap out the broken block with a direct url file streaming download mechanism
new_block = """    import urllib.request
    from transformers import GPT2TokenizerFast

    os.makedirs(cfg.data_dir, exist_ok=True)
    tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

    def encode_split(split_name: str) -> int:
        # Stream the source data files directly from the master repository files
        repo_url = "https://raw.githubusercontent.com/pytorch/examples/main/word_language_model/data/wikitext-2"
        file_target = "valid.txt" if split_name == "val" else "train.txt"

        print(f"  Downloading wikitext-2 raw split: {split_name}...")
        with urllib.request.urlopen(f"{repo_url}/{file_target}") as response:
            texts = response.read().decode('utf-8')

        ids = tokenizer.encode(texts)"""

# Write the patched script out to the file system
content = content.replace(old_block, new_block)
train_py_path.write_text(content)
print("✓ Successfully patched train.py to use direct stream downloads!")
```

```python
import re
from pathlib import Path

train_py_path = Path("train.py")
content = train_py_path.read_text()

# Target the original broken Hugging Face block inside train.py
old_block = """    from datasets     import load_dataset
    from transformers import GPT2TokenizerFast

    os.makedirs(cfg.data_dir, exist_ok=True)
    tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")
    dataset   = load_dataset("wikitext", "wikitext-2-raw-v1")

    def encode_split(split_name: str) -> int:
        hf_split = "validation" if split_name == "val" else split_name
        texts    = "\\n\\n".join(
            t for t in dataset[hf_split]["text"] if t.strip()
        )
        ids = tokenizer.encode(texts)"""

# Define the direct web stream patch
new_block = """    import urllib.request
    from transformers import GPT2TokenizerFast

    os.makedirs(cfg.data_dir, exist_ok=True)
    tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

    def encode_split(split_name: str) -> int:
        # Stream the source data files directly from the master repository files
        repo_url = "https://raw.githubusercontent.com/pytorch/examples/main/word_language_model/data/wikitext-2"
        file_target = "valid.txt" if split_name == "val" else "train.txt"

        print(f"  Downloading wikitext-2 raw split: {split_name}...")
        with urllib.request.urlopen(f"{repo_url}/{file_target}") as response:
            texts = response.read().decode('utf-8')

        ids = tokenizer.encode(texts)"""

# Apply the text swap and save the file
content = content.replace(old_block, new_block)
train_py_path.write_text(content)
print("✓ Successfully patched train.py to use direct stream downloads!")
```

```python
import re

with open("/content/eval_extrap.py") as f:
    src = f.read()

# Fix 1: weights_only for PyTorch 2.6
src = src.replace(
    "ckpt = torch.load(ckpt_path, map_location=device)",
    "ckpt = torch.load(ckpt_path, map_location=device, weights_only=False)"
)

# Fix 2: strip private/non-dataclass keys before passing to get_config
src = src.replace(
    "cfg = get_config(**{\n            k: v for k, v in saved_cfg.items()\n            if k != \"seq_len\"   # we'll override per eval\n        })",
    "cfg = get_config(**{\n            k: v for k, v in saved_cfg.items()\n            if k != \"seq_len\" and not k.startswith(\"_\")\n        })"
)

with open("/content/eval_extrap.py", "w") as f:
    f.write(src)

# Verify both fixes applied
print("weights_only fix:", "weights_only=False" in src)
print("_pe_instance fix:", 'not k.startswith("_")' in src)
```

```python
# Fix the RoPE interpolation bug
with open("/content/pos_encoding_variants.py") as f:
    src = f.read()

src = src.replace(
    "t     = torch.arange(length, device=device).float() / scale",
    "t     = torch.arange(length, device=device).float() * scale"
)

with open("/content/pos_encoding_variants.py", "w") as f:
    f.write(src)

print("fix applied:", "* scale" in src)
```

```python
with open("/content/eval_extrap.py") as f:
    src = f.read()

src = src.replace(
    'other_issues = [k for k in missing + unexpected if "pos_encoding" not in k]',
    'other_issues = [k for k in missing + unexpected if "pos_encoding" not in k and "_pe_module" not in k]'
)

with open("/content/eval_extrap.py", "w") as f:
    f.write(src)

print("filter fix applied")
```
